Прочитаем данные из train.csv. Т.к. файл очень большой и не помещается полностью в память, то прочитаем 5000000 строк

```python
chunk_size = 5000000
chunks = []
df_train = pd.DataFrame()
reader = pd.read_csv("drive/MyDrive/DataSet/train.csv", chunksize=chunk_size)

for chunk in reader:
  df_train = pd.concat([df_train, chunk], ignore_index=True)
  break

df_train.head()
```

Убирем из dataframe'а лекции. Они имеют значение 1 в колонке content_type_id

```python
df_no_lectures = df_train.drop(df_train[df_train['content_type_id'] == 1.000000].index)
```

Проверим на наличие пустых строк в df_no_lectures

```python
df_no_lectures.isnull().sum()
```

У нас есть поле prior_question_elapsed_time - Среднее время в миллисекундах, которое потребовалось пользователю для ответа на каждый вопрос из предыдущей группы вопросов, игнорируя любые промежуточные лекции. Имеет значение NULL для первого пакета вопросов или лекции пользователя. Также у нас есть поле prior_question_had_explanation - означающее увидел ли пользователь объяснение и правильный ответ после ответа на предыдущий блок вопросов, игнорируя любые лекции между ними. Значение распределяется по одному пакету вопросов и равно null для первого пакета вопросов или лекции пользователя. Обычно первые несколько вопросов, которые видит пользователь, были частью вводного диагностического теста, по которому он не получил никакой обратной связи.

Соответственно мы можем разделить наш dataframe на 2 - первый будет содержать данные, где prior_question_elapsed_time = null, а второй - все остальные

```python
df_no_prior_question = df_no_lectures[df_no_lectures['prior_question_elapsed_time'].isnull()]
df_no_lectures = df_no_lectures.drop(df_no_lectures[df_no_lectures['prior_question_elapsed_time'].isnull()].index)
```

Проверим, что в df_no_lectures не осталось null

```python
df_no_lectures.isnull().sum()
```

Посмотрим, есть ли корреляция между answered_correctly и другими полями в df_no_lectures

```python
df_no_lectures.corr()
```

Визуализируем эту историю

```python
import matplotlib.pyplot as plt

plt.matshow(df_no_lectures.corr())
plt.show()
```
